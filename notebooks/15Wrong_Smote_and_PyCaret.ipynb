smote 1

```python
X = df_smote_1.drop(['TenYearCHD'], axis = 1)
Y = df_smote_1['TenYearCHD']

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, random_state = 42, stratify = Y, shuffle = True)
     

# Handling Imbalanced Dataset
smote = SMOTE(random_state = 42)
X_smote, Y_train_final = smote.fit_resample(X_train, Y_train)

X_smote.reset_index(drop=True, inplace=True)
X_test.reset_index(drop=True, inplace=True)
Y_train_final.reset_index(drop=True, inplace=True)
Y_test.reset_index(drop=True, inplace=True)

X_train_final_df = pd.DataFrame(X_smote, columns=X_train.columns)
X_test_final_df = pd.DataFrame(X_test, columns=X_test.columns)

# Combine the dataframes vertically
X_combined = pd.concat([X_train_final_df, X_test_final_df], axis=0, ignore_index=True)
y_combined = pd.concat([Y_train_final, Y_test], axis=0, ignore_index=True)

df_smote_1d = pd.concat([X_combined, y_combined], axis=1, ignore_index=True)

df_smote_1d = df_smote_1d.rename(columns=dict(zip(df_smote_1d.columns, df_without_smote.columns)))
df_smote_1d = df_smote_1d.rename(columns={11: "TenYearCHD"})
```

```python
from pycaret.classification import *

# Initialize the setup
clf = setup(data=df_smote_1d, target='TenYearCHD')

# Compare models
best_model = compare_models(sort='AUC')
```

### Smote 4: Undersampling 0.3

```python
X = df_smote_4.drop(['TenYearCHD'], axis = 1)
Y = df_smote_4['TenYearCHD']

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, random_state = 42, stratify = Y, shuffle = True)
     
from imblearn.under_sampling import RandomUnderSampler
# Handling Imbalanced Dataset
undersampler = RandomUnderSampler(sampling_strategy=0.3, random_state=42)
X_smote, Y_train_final = undersampler.fit_resample(X_train, Y_train)

X_smote.reset_index(drop=True, inplace=True)
X_test.reset_index(drop=True, inplace=True)
Y_train_final.reset_index(drop=True, inplace=True)
Y_test.reset_index(drop=True, inplace=True)

X_train_final_df = pd.DataFrame(X_smote, columns=X_train.columns)
X_test_final_df = pd.DataFrame(X_test, columns=X_test.columns)

# Combine the dataframes vertically
X_combined = pd.concat([X_train_final_df, X_test_final_df], axis=0, ignore_index=True)
y_combined = pd.concat([Y_train_final, Y_test], axis=0, ignore_index=True)

df_smote_4d = pd.concat([X_combined, y_combined], axis=1, ignore_index=True)

df_smote_4d = df_smote_4d.rename(columns=dict(zip(df_smote_4d.columns, df_without_smote.columns)))
df_smote_4d = df_smote_4d.rename(columns={11: "TenYearCHD"})
```

### Smote 5: 0.2 oversample and 0.5 undersample

```python
from imblearn.over_sampling import SMOTE
from imblearn.under_sampling import RandomUnderSampler
from imblearn.pipeline import make_pipeline

X = df_smote_5.drop(['TenYearCHD'], axis = 1)
Y = df_smote_5['TenYearCHD']

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, random_state = 42, stratify = Y, shuffle = True)


# Define the SMOTE and RandomUnderSampler objects with the desired sampling strategies
smote5 = SMOTE(sampling_strategy=0.2, random_state=42)
undersampler5 = RandomUnderSampler(sampling_strategy=0.5, random_state=42)

# Create a pipeline combining SMOTE and RandomUnderSampler
smote5 = make_pipeline(smote5, undersampler5)

# Apply the pipeline to your data
X_smote, Y_train_final = smote5.fit_resample(X_train, Y_train)

X_smote.reset_index(drop=True, inplace=True)
X_test.reset_index(drop=True, inplace=True)
Y_train_final.reset_index(drop=True, inplace=True)
Y_test.reset_index(drop=True, inplace=True)

X_train_final_df = pd.DataFrame(X_smote, columns=X_train.columns)
X_test_final_df = pd.DataFrame(X_test, columns=X_test.columns)

# Combine the dataframes vertically
X_combined = pd.concat([X_train_final_df, X_test_final_df], axis=0, ignore_index=True)
y_combined = pd.concat([Y_train_final, Y_test], axis=0, ignore_index=True)

df_smote_5d = pd.concat([X_combined, y_combined], axis=1, ignore_index=True)

df_smote_5d = df_smote_5d.rename(columns=dict(zip(df_smote_5d.columns, df_without_smote.columns)))
df_smote_5d = df_smote_5d.rename(columns={11: "TenYearCHD"})
```

```python
from pycaret.classification import *

# Initialize the setup
clf = setup(data=df_smote_5d, target='TenYearCHD')

# Compare models
best_model = compare_models(sort='AUC')
```
